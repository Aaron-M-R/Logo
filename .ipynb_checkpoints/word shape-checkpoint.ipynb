```python

# Loading in the files
data_dict = dict()
n = 1
for i in shape_files:
    data_dict[n] = i
    n += 1

# Finding the best method
image_a = list()
image_b = list()
similarity = list()
methods = list()


# Double loop through every word image
for numa, imagea in data_dict.items():
    for numb, imageb in data_dict.items():
        for method in [1,2,3]:
            image_a.append(filea)
            image_b.append(fileb)
            score = cv2.matchShapes(get_contours(imagea),
                                    get_contours(imageb),
                                    method,0)
            similarity.append(score)
            methods.append(method)

wsh_df = pd.DataFrame({'Image A':image_a,
                        'Image B':image_b,
                        'Score':similarity,
                        'Method':methods})


# Analyzing the methods
wsh_df = wsh_df.sort_values(by=['Method','Score'], ascending=[True, True])

for i in wsh_df['Method'].unique():
    temp_df = wsh_df.loc[wsh_df['Method']==i]
    pivot_df = temp_df.pivot('Image A',columns='Image B')[['Score']]
    sns.heatmap(pivot_df, annot=True).set_title(f'Method {i} Heatmap')
    plt.show()
```

```python
# Setting the function to load the contours
def get_contours(input_image, show_image=False):
#     _, binary = cv2.threshold(input_image, 225, 255, cv2.THRESH_BINARY_INV)
#     binary = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
    
    binary = cv2.cvtColor(np.array(input_image, dtype=np.uint8), cv2.COLOR_BGR2GRAY)
    inverted = 255-binary
    contours, hierarchy = cv2.findContours(inverted,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
#     print(contours)
    
#     image = cv2.drawContours(np.array(input_image, dtype=np.uint8), contours, -1, (0, 255, 0), 2)
    
#     if show_image == True:
#         plt.imshow(image)
#         plt.show()
        
#         plt.imshow(binary, cmap="gray")
#         plt.show()
    
    return contours
```

```python
from PIL import Image, ImageDraw, ImageFont
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Finding the best method
image_a = list()
image_b = list()
similarity = list()
methods = list()

data_dict = dict()


words = ['Hello', 'Mello', 'Trickster', 'Pleasant', 'Unbelievable']

for word in words:
    data_dict[word] = get_shape_image(word)




# Double loop through every word image
for filea, imagea in data_dict.items():
    for fileb, imageb in data_dict.items():
        for method in [3]:
            image_a.append(get_contours(imagea)[0])
            image_b.append(get_contours(imageb)[0])
            score = 1 - cv2.matchShapes(imagea,
                                    imageb,
                                    method,0)
            similarity.append(score)
            methods.append(method)
        
wsh_df = pd.DataFrame({'Image A':image_a,
                        'Image B':image_b,
                        'Score':similarity,
                        'Method':methods})

# Analyzing the methods
wsh_df = wsh_df.sort_values(by=['Method','Score'], ascending=[True, True])

for i in wsh_df['Method'].unique():
    temp_df = wsh_df.loc[wsh_df['Method']==i]
#     pivot_df = temp_df.pivot('Image A',columns='Image B')[['Score']]
#     sns.heatmap(pivot_df, annot=True).set_title(f'Method {i} Heatmap')
    plt.show()
    
```
